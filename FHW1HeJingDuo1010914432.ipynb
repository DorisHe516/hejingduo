#### 7. Missing data can be considered "across rows" or "down columns".  Consider how `df.dropna()` or `del df['col']` should be applied to most efficiently use the available non-missing data in your dataset and briefly answer the following questions in your own words

1. Provide an example of a "use case" in which using `df.dropna()` might be peferred over using `del df['col']`<br><br>

```python
import pandas as pd

data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, None, 30, 40],
        'Score': [85, 90, None, 75]}

abc = pd.DataFrame(data)
print(abc)
```

```python
abc.dropna()
```

In this data, if we use "df.dropna()", we will only delete missing values, but if we use "del df['col']", we will delete an entire column of data. Therefore, it can be seen that using d in this situation can reduce data loss.

2. Provide an example of "the opposite use case" in which using `del df['col']` might be preferred over using `df.dropna()` <br><br>

```python
import pandas as pd


data = {'Name': ['Alice', 'Bob', 'Charlie', 'David'],
        'Age': [25, 30, 35, 40],
        'Gender': ['F', 'M', 'M', 'M'],
        'Address': [None, None, None, None]} 
bbb = pd.DataFrame(data)
print(bbb)
```

```python
del bbb["Address"]
print(bbb)
```

In this data, we do not have missing values, only useless values named None. If we use "bbb.dropna()", this does not remove any useless values, but if we use "del bbb['col']", we can precisely remove the useless values in the "Address" column.

3. Discuss why applying `del df['col']` before `df.dropna()` when both are used together could be important<br><br>

If a set of data contains a large number of missing values, using d directly may result in the loss of a significant amount of valid data. From this, we can conclude that we should first use "del bbb["Address"](del df['col'])" to precisely delete irrelevant columns, and then use "bbb.dropna()(df.dropna())" to delete the remaining missing values.
